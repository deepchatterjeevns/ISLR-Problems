## Part b

```python
g = sns.PairGrid(df)
g.map_upper(plt.scatter, s=3)
g.map_diag(plt.hist)
g.map_lower(plt.scatter, s=3)
g.fig.set_size_inches(12, 12)
```

Average number of rooms per dwelling (rm) apears to have a strong positive linear relationship with price. Also, % lower status of the population(LSTAT) apears to have a strong non-linear negative relationship with price.
